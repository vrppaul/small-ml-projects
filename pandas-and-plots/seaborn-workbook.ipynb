```python
import pandas as pd
reviews = pd.read_csv('data/wine-reviews/winemag-data_first150k.csv', index_col=0)
import seaborn as sns
```

```python
sns.countplot(reviews.points)
```

```python
sns.kdeplot(reviews.query('price < 200').price)
```

Comparison to Pandas line()

```python
(reviews
     .query('price < 200')
     .price
     .value_counts()
     .sort_index()
     .plot.line())
```

```python
sns.kdeplot(reviews
               .query('price < 200')
               .price
               .dropna()
               .sample(5000), 
           reviews
               .query('price < 200')
               .points
               .dropna()
               .sample(5000))
```

```python
sns.distplot(reviews.points, bins = 10, kde = False)
```

```python
sns.jointplot(x='price', y='points', data=reviews.query('price<100'))
```

```python
sns.jointplot(x='price', y='points', data=reviews.query('price < 100'), kind='hex', gridsize=20)
```

```python
df = reviews[reviews.variety.isin(reviews.variety.value_counts().head(5).index)]
sns.boxplot(
    x='variety',
    y='points',
    data=df)
```
